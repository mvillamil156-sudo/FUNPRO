```python
password = input("Ingrese su contraseña: ")
mayusculas = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
minusculas = "abcdefghijklmnñopqrstuvwxyz"
especiales = "~!@#$%^&*()_+"
numero = "0123456789"
```

```python
c1=0
if password[0] in mayusculas:
    c1=1
    print("Has cumplido la primera condición")
```

```python
c2=0
a=0
for letra in password:
    if letra in numero:
        a=a+1
if a>=2:
    c2=1
    print("Has cumplido la segunda condición")
```

```python
c3=0
if len(password) >= 10:
    c3=1
    print("Has cumplido la tercera condición")
```

```python
c4=0
for letra in password:
    if letra in minusculas:
        c4=1
        print("Has cumplido la cuarta condición")
        break
```

```python
c5=0
for esp in especiales:
    if esp in password:
        c5=1
        print("Has cumplido la quinta condición")
        break
```

```python
if c1==1 and c2==1 and c3==1 and c4==1 and c5==1:
    print("La contraseña cumple todas las condiciones.Tienes acceso.")
else:
    print("No tienes acceso, te faltan los siguientes requisitos:")
    if c1==0:
        print("-La contraseña debe empezar con una letra mayúscula.")
    if c2==0:
        print("-Debe contener al menos dos números.")
    if c3==0:
        print("-Debe tener mínimo 10 caracteres.")
    if c4==0:
        print("-Debe incluir al menos una letra minúscula.")
    if c5==0:
        print("-Debe tener al menos un carácter especial (~!@#$%^&*()_+).")
```
