```python
ahorro=int(input("¿Cuánto dinero tengo?"))
opc=input("¿Deseo comprar algo?")
while ahorro>0 and opc=="s":
 precio=int(input("¿Qué cuesta?"))
 if precio<=ahorro:
  print("Voy a comprar más")
  ahorro=ahorro-precio #acumulador
 else:
  print("No me alcanza")
opc=input("¿Deseo comprar algo?")
```

```python
num=int(input("Ingresar número"))
while num>=1:
  print("Tiene cubo")
  cubo=num*num*num
  print(f"Este es el nuevo resultado {cubo}")
if num<=0:
  print("No es posible")
```
